```python
import pandas as pd
import numpy as np
```

```python
green_data = pd.read_csv("greenland_mass_200204_202102.txt" , skiprows=31, sep='\s+', header=None)
ant_data = pd.read_csv("antarctica_mass_200204_202102.txt" , skiprows=31, sep='\s+', header=None)
sea_level_data = pd.read_excel('Sea_Level_Data3.xls')

display(green_data)
display(ant_data)
display(sea_level_data)
```

```python
data = [green_data[0], green_data[1], ant_data[1]]

headers = ["Date", "Greenland", "Antarctica"]

total_df = pd.concat(data, axis=1, keys=headers)

print(total_df)
```

```python
### picking out data from sea level data

data1 =[sea_level_data['Column3'], sea_level_data['Column11']]

headers1 =["Date", "Mean Sea Level"]

new_df = pd.concat(data1, axis=1, keys=headers1)

print(new_df)
```

```python
from datetime import datetime

def fraction2datetime(year_fraction: float) -> datetime:
    year = int(year_fraction)
    fraction = year_fraction - year
    first = datetime(year, 1, 1)
    aux = datetime(year + 1, 1, 1)
    return first + (aux - first)*fraction
```

```python
new_df["Date"] = new_df["Date"].apply(fraction2datetime)
total_df["Date"] = total_df["Date"].apply(fraction2datetime)   
    
print(new_df)
print(total_df)
```

```python
merged_df = pd.merge_asof(
    total_df, new_df, on="Date", tolerance=pd.Timedelta(days=30), direction="nearest"
)

pd.set_option('display.max_rows', None)
display(merged_df)
```

```python
n=3

merged_df.drop(merged_df.tail(n).index,
        inplace = True)
```
